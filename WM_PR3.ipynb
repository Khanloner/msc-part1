```python
import networkx as nx
import pylab as plt
```

```python
A = nx.DiGraph()
```

```python
A.add_weighted_edges_from([('A','B',1),('A','C',1),('B','C',1)])
print(nx.pagerank(A))
```

```python
# Compute pagerank
pagerank_scores = nx.pagerank(A,alpha = 0.85) #alphha is the damping factor
print(pagerank_scores)
```

```python
nx.draw(A,with_labels=True)
```
