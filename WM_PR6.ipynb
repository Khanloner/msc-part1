```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import csv
```

```python
def crawl(url, depth = 2, visited=None, csv_writer = None, parent_url=None):
    if visited is None:
        visited = set()
    if depth == 0 or url in visited:
        return
    print(f'Crawling: {url} (Found on:{parent_url})')
    visited.add(url)
    if csv_writer:
        csv_writer.writerow([url, parent_url])
    try:
        response = requests.get(url, timeout=5)
        if response.status_code !=200:
            return
        soup = BeautifulSoup(response.text,"html.parser")
        links = [urljoin(url, a['href']) for a in soup.find_all("a",href=True)]

        for link in links:
            if link.startswith("http") or link.startswith("https"):
                crawl(link,depth-1, visited,csv_writer,url)
    except requests.RequestException as e:
        print(f'Error: {e}')
```

```python
start_url = "https://www.scrapingbee.com/blog/crawling-python/"
with open("crawled_links.csv","w",newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["URL","Found On"])
    crawl(start_url,depth=2,csv_writer=writer, parent_url="root")
```
