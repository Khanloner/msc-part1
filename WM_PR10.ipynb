```python
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk

nltk.download("vader_lexicon")
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("punkt_tab")


sample_texts = [
    "I absolutely love this product! It's fantastic and works like a charm.",
    "This is the worst experience I've ever had. So disappointing!",
    "The service was okay, nothing special but not terrible either.",
    "I am feeling really sad and lonely today.",
    "Wow! This is amazing, I couldn't be happier!",
    "Meh, it's just an average product. Nothing too exciting."
]
```

```python
def analyze_sentiment_textblob(text):
    """Analyze sentiment using textblob"""
    blob = TextBlob(text)
    sentiment_score = blob.sentiment.polarity

    if sentiment_score>0:
        sentiment = "Positive"
    elif sentiment_score<0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return sentiment, sentiment_score


def analyze_sentiment_vader(text):
    """Analyze sentiment using VADER (best for short, social media like text)"""
    sia = SentimentIntensityAnalyzer()
    sentiment_score = sia.polarity_scores(text)["compound"]

    if sentiment_score>0:
        sentiment = "Positive"
    elif sentiment_score<0:
        sentiment = "Negative"
    else:
        sentiment = "Neutral"
    return sentiment, sentiment_score
```

```python
for text in sample_texts:
    textblob_sentiment, textblob_score = analyze_sentiment_textblob(text)
    vader_sentiment, vader_score = analyze_sentiment_vader(text)

    print(f"Text: {text}")
    print(f" Textblob Sentiment: {textblob_sentiment} (Score: {textblob_score})")
    print(f" VADER Sentiment: {vader_sentiment} (Score: {vader_score})")
    print("-"*80)
```
