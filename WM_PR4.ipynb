```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
    ["Milk","Bread","Butter"],
    ["Milk","Diaper","Juice","Bread"],
    ["Milk","Diaper","Bread"],
    ["Diaper","Juice","Bread"],
    ["Milk","Diaper","Juice","Bread"]
]
```

```python
# Extract all items
all_items = set()
for transaction in dataset:
    for item in transaction:
        if item not in all_items:
            all_items.add(item)
print("All Items ", all_items)
```

```python
transaction_list = []

for transaction in dataset:
    transaction_dict={}

    for item in all_items:
        transaction_dict[item] = item in transaction
    transaction_list.append(transaction_dict)

print(transaction_list)
df = pd.DataFrame(transaction_list)
```

```python
df
```

```python
frequent_itemsets = apriori(df, min_support = 0.6, use_colnames = True)
print("Frequent Itemset:")
frequent_itemsets
```

```python
# rules = association_rules(frequent_itemsets, metric="confidence",min_threshold=0.7, num_itemsets = 2)
rules = association_rules(
    frequent_itemsets,
    metric="confidence",
    min_threshold=0.7
)
print("Association Rules:")
rules
```

```python
strong_rules = rules[(rules['confidence']>0.7) & (rules['lift']>1) ]
print("Strong Association rules (confidence > 0.7 and Lift>1): ")
strong_rules.head()
```
