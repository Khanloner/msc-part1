```python
import os
import pandas as pd
from datetime import date
today = date.today()
end_date = today

search_term = 'Bhavesh Bhatt Data Scientist'
from_date = '2019-01-01'
```

```python
os.system(
    f"snscrape --since {from_date} twitter-search "
    f"'{search_term} until:{end_date}' > result-tweets.txt"
)

if os.stat("result-tweets.txt").st_size == 0:
    counter = 0
else:
    df = pd.read_csv("result-tweets.txt", names = ['link'])
    counter = df.size
print("Number of tweets: "+str(counter))
```

```python
max_results = 100
```

```python
extracted_tweets = (
    f"snscrape --format '{{content!r}}' "
    f"--max-results {max_results} "
    f"--since {from_date} "
    f"twitter-search '{search_term} until:{end_date}' "
    f"> extracted-tweets.txt"
)

print(extracted_tweets)
os.system(extracted_tweets)
print(os.path.exists("extracted-tweets.txt"))
print(os.stat("extracted-tweets.txt").st_size)


os.system(extracted_tweets)
if os.stat("extracted-tweets.txt").st_size == 0:
    print('No Tweets found')
else:
    df = pd.read_csv('extracted-tweets.txt', names=['content'])
    for row in df['content'].iteritems():
        print(row)
```
