```python
# WM_PR9
import os
from itertools import combinations
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def read_file(file_path):
    with open(file_path,"r", encoding="utf-8") as file:
        return file.read()

def get_text_files():
    return [f for f in os.listdir() if f.endswith(".txt")]
```

```python
def compare_documents(file1, file2,documents):
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([documents[file1], documents[file2]])
    return cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])[0][0]
```

```python
def check_plagiarism():
    text_files = get_text_files()
    documents = {file: read_file(file) for file in text_files}

    for file1, file2 in combinations(text_files,2):
        similarity = compare_documents(file1, file2, documents)
        print(f"Plagiarism Score: {(file1, file2, f'{round(float(similarity) * 100,2)}%' )}")
```

```python
check_plagiarism()
```

```python
# Alternate code
def check_plagiarism():
    text_files = get_text_files()
    if len(text_files) < 2:
        print("Not enough .txt files to compare.")
        return

    documents_list = [read_file(f) for f in text_files]
    
    # Fit on ALL documents at once to create a master vocabulary
    vectorizer = TfidfVectorizer()
    try:
        tfidf_matrix = vectorizer.fit_transform(documents_list)
    except ValueError as e:
        print(f"Error: {e}. Check if your files are empty.")
        return

    # Compare each pair using the pre-calculated matrix
    for i, j in combinations(range(len(text_files)), 2):
        similarity = cosine_similarity(tfidf_matrix[i], tfidf_matrix[j])[0][0]
        file1, file2 = text_files[i], text_files[j]
        print(f"Plagiarism Score: {file1} vs {file2} -> {round(float(similarity) * 100, 2)}%")
```

```python
check_plagiarism()
```
